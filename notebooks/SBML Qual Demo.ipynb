```python
import requests

from mira.sources.biomodels import get_sbml_model
from mira.sources.sbml.qual_api import template_model_from_sbml_qual_string

MODEL_URL = "https://git-r3lab.uni.lu/covid/models/-/raw/master/Executable%20Modules/SBML_qual_build/sbml/Orf3a_stable.sbml?ref_type=heads"
BIOMODEL_ID = "BIOMD0000000562"
```

## We load in biomodel BIOMD0000000562 from the biomodel's database and convert it to a semantically similar MIRA template model

```python
model_text = get_sbml_model(BIOMODEL_ID)
tm = template_model_from_sbml_qual_string(model_text)
```

## Identify all concepts loaded in from the biomodel

Qualitative species' annotations are not provided for models on the biomodel's website

```python
for initial in tm.initials:
    print(initial)
```

## Qualitative models do not have parameters

```python
tm.parameters
```

## Visualize the template model structure loaded in from a SBML Qual document from the biomodels website

```python
tm.draw_jupyter()
```

## We then convert an SBML Qual document from the Covid-19 disease map

The specific file is here: https://git-r3lab.uni.lu/covid/models/-/blob/master/Executable%20Modules/SBML_qual_build/sbml/Orf3a_stable.sbml?ref_type=heads


```python
model_text = requests.get(MODEL_URL).text
tm = template_model_from_sbml_qual_string(model_text)
tm.draw_jupyter()
```

## We identify all concepts and their identifiers from the model that is apart of the Covid-19 disease map

Qualitative species' annotations have been provided for most of the species' in the model

```python
for initial in tm.initials.values():
    print(f"{initial.concept.display_name} | {initial.concept.identifiers}")
```
