### Prompting development for MIRA equations extractions

This notebook outlines ideas for prompting for extracting mathematical equations from PDFs in the MIRA framework.

There are clearly separable problem areas, which will be better managed by detailing and resolving the prompting. An agenda based approach will systematically address extraction challenges by organizing the process into distinct agenda items, each targeting specific aspects of the process:

1. **Execution error handling**
    - automatically alert users when execution errors occur
    - implement automatic redo functionality for failed attempts

2. **Time dependency classification**
    - provide clear classification of time-dependent variables
    - distinguish non time-dependent variables from parameters
    - ensure proper categorization in time dependency checks

3. **Parameter usage consistency check**
    - detect missed parameters that should be included
    - identify duplicated parameters across equations
    - flag unused parameters in the extraction
    - check super- and subscript usage consistency
    
4. **Equation structure validation**
    - check if extracted equations are mathematically coherent 
    - verify structural integrity of mathematical expressions

5. **Diagnostics and correctness evaluation**
    - identify possible issues during processing
    - assign a correctness score after validation
    - provide warnings about detected problems

Since each agent requires a distinct approach and prompt configuration, the LLM can achieve better focus (rather than receiving a summarized, less detailed message).

---
Other possible agenda items:

6. Symbol Validation – Are all variables and parameters defined? this focuses more on JSON

7. Biological Context Tagging – Are compartments semantically labeled (e.g., S = susceptible)?

8. JSON Structure Integrity – Is the output JSON consistent and complete?
